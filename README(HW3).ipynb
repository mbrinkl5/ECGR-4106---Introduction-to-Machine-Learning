# ECGR 4106/5106 — Homework 3

Sequence-to-sequence English/French machine translation with GRU encoder-decoder models, in PyTorch.

## Contents

- `ECGR4106_HW_3.ipynb` — full notebook (all three problems, training, plots, and attention maps)
- `vast_english_french.txt` — dataset (555 English-French sentence pairs)

## Overview

- **Problem 1:** Baseline GRU encoder-decoder, English to French. Train/validation loss curves, exact-match sequence accuracy, corpus BLEU-4, and qualitative samples.
- **Problem 2:** Previous model with Bahdanau-style attention. Loss curves, both metrics, comparison to Problem 1, and attention-map visualizations.
- **Problem 3:** Reversed direction (French to English) with both models, loss curves, both metrics, qualitative samples, and a synthesis comparing the two directions.

## Setup

Trained on Google Colab (T4 GPU). Word-level translation. Each model is an embedding + GRU encoder with a GRU decoder (baseline) or a Bahdanau-attention decoder. SGD optimizer with loss applied to LogSoftmax outputs, cross-entropy (NLL) loss, teacher-forcing ratio 1.0, hidden size 256, dropout 0.1 in the attention decoder, 50 epochs, batch size 1. The dataset is split 80/20 (444 train / 111 validation) once and reused across all problems. BLEU-4 is reported as the average sentence-level BLEU-4 score computed with NLTK's sentence_bleu() function using Method-1 smoothing.

## Results (validation)

| Model | Direction | BLEU-4 | Seq accuracy | Final val loss |
|---|---|---|---|---|
| Baseline | EN → FR | 0.0317 | 0.00% | 4.7632 |
| Attention | EN → FR | 0.0717 | 0.00% | 4.3829 |
| Baseline | FR → EN | 0.333 | 0.00% |  4.6195 |
| Attention | FR → EN | 0.0657 | 0.00% | 4.3675 |

The attention mechanism reduced the validation loss and improved BLEU-4 scores in both translation directions. Although French-to-English achieved the lowest validation loss, English-to-French produced higher BLEU-4 scores, indicating better translation quality overall. On a dataset of only 555 sentence pairs, both models exhibited overfitting and achieved 0.00% exact-match accuracy, making the qualitative translation examples and BLEU-4 scores more informative measures of performance.
